```python
# Kkma import 하기
# 꼬꼬마 라이브러리 라고 부름
from konlpy.tag import Kkma
```

```python
km = Kkma()
```

```python
km.nouns(u'지금은 오후 5시 58분 입니다')
```

```python
km.morphs('공부를 하면할수록 모르는게 많다는 것을 알게 됩니다.')
```

```python
km.pos('다 까먹어버렸네요?ㅋㅋㅋㅋ')
```

```python
km.sentences('곧, 집에 갈 시간이예요.')
```
