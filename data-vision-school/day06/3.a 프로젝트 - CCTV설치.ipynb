```python
# 합치는 것, 중요함!
# on= 파라미터로 공통되는 컬럼을 기준으로 삼는다 
# 비슷한 함수로는 join(), concat() 등이 있음, 성격이 조금씩 다름
# 만약 공통되는 컬럼이나 컬럼 이름이 다를 때는, rename()으로 먼저 컬럼명을 통일 시켜주면 됨
merged_pop_cctv = pd.merge(pop_Seoul, sel_cctv, on='구별')
```
