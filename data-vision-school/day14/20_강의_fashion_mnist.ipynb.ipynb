Feature Scaling

학습이 잘 되도록 0과 1사이의 값으로 노말라이징(normalizing) 한다.

```python
#최대값이 255이므로 255로 나누어 주면 노말라이징이 된다 255가 1이 됨
#실수처리 하기 위해ㅓ서 255.0 
```

```python
training_images = training_images / 255.0 
```

이미지의 가로 세로를 전부 일렬로 만드는 작업이 flatten 이다.

모델을 만들어 보자.

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense, Flatten
```

```python
# import를 안해서 tf에서 가져옴
model = tf.keras.models.Sequential()
```

```python
# 데이터 입력 부분을 사진 행렬을 일자로 만들어 준다
# Dense()가 아닌 Flatten()을 사용
model.add( tf.keras.layers.Flatten() )
```

```python
model = Sequential()
```

```python
model.add( Flatten() )
```

Flatten()
이미지 행렬 예줌  
0 1 2 0 3 1 0  
2 3 5 1 3 1 0  
9 1 0 2 3 9 0  
2차원 이미지 를 행으로 쭈욱 평평하게 연결해주는 것   (아래처럼) 하나의 행으로 만들어 줌  
0 1 2 0 3 1 0 2 3 5 1 3 1 0 9 1 0 2 3 9 0  


28 x 28 784

```python
model.add( Dense(units= 128, activation=tf.nn.relu ))
```

마지막 unit은 10개이다, 1개가 아님


```python
# output layer가 10개
# 3개 이상의 분류는 해당 분류수 만큼 units= 를 같게 지정해준다
# 이번프로젝트의 분류 10개임
# 액티베이션 함수로 소프트맥스를 사용

model.add( Dense(units= 10, activation=tf.nn.softmax))
```

소프트맥스   
0  
1  

0 과 1사이로 결과를   
결과의 총 합이 1이 되게 만드는 액티베이션 함수  

우리는 그 결과중에 가장 높은 수를 찾으면 됨

**Sequential**: SEQUENCE layers 로 모델링

**Flatten**: 2차원, 3차원을 1차원으로 만드는것

**Dense**: 뉴런을 레이어에 붙인다.

각 레이어는 **activation function** 이 필요하다. 

**Relu** effectively means "If X>0 return X, else return 0" -- so what it does it it only passes values 0 or greater to the next layer in the network.

**Softmax** 여러개의 값 중에서 가장 큰 값을 선택. [0.1, 0.1, 0.05, 0.1, 9.5, 0.1, 0.05, 0.05, 0.05], 여기서 가장 큰 값을 1로 만들고 나머지는 0으로 만들어준다. [0,0,0,0,1,0,0,0,0] 


optimizer and loss function 으로 컴파일 한 후에  **model.fit** 함수로 학습한다.. 

```python
# sparse_categorical_crossentropy 3가지 이상 분류잉ㄹ 때 
# metrics= ['accuracy'] 분류이므로 accuray도 넣어 준다
```

```python
#** 중요!!y 값이 정수로 되어 있을 때 사용할 수 있음

model.compile(optimizer=tf.optimizers.Adam(), loss = 'sparse_categorical_crossentropy', metrics= ['accuracy'] )
```

```python
training_images.shape
#3차원 
```

```python
model.fit(training_images, training_labels, epochs=5)
```

accuracy 가 약 90% 라는 뜻. 5 epochs 라 금방 끝났다.

## Exercise 7: 

accuracy 나 loss 가, 우리가 원하는 특정 값이 되면 자동으로 학습을 멈추가 하고 싶다. 즉, 특정값에 도달하면 학습을 멈추게 할 수 있는 콜백기능을 제공한다. callbacks

```python
model_7 = Sequential()
model_7.add( Flatten() )
model_7.add( Dense(units= 512, activation=tf.nn.relu ))
model_7.add( Dense(units= 10, activation=tf.nn.softmax))
model_7.compile(optimizer=tf.optimizers.Adam(), loss = 'sparse_categorical_crossentropy', metrics= ['accuracy'] )
```

callbacks

callbacks는 내가 실행하는 것이 아니고 어떤 조건이 되었을 때 callbacks 함수가 작동을 하면서 내가 만든 사용자함수를 호출 (실행)시켜주는 것  
(내가 사용한 적은 없음, 만들기만 했지만, 텐서플로우에서 자동으로 실행하는 것: callbacks)

-사용법-
내 클래스를 만들고 (상속을 받을 수 있게 함tk.keras.callbacks.Callback)  
클래스 내의 메소드는 정해진 규칙을 따른다  
def on_epoch_end(self, epoch, logs={} ):  
그 다음 조건은 내가 만들면 됨

[참고: 텐서플로우.org](https://www.tensorflow.org/guide/keras/custom_callback)

```python
class myCallback(tf.keras.callbacks.Callback):
  
  def on_epoch_end(self, epoch, logs={} ):
    #if (logs.get('loss') < 0.4 ):
    if (logs.get('accuracy') > 0.90 ):
      #print("\n오차가 0.4 미만이라 학습을 멈춥니다.")
      print("\n정확도 0.9 이상이면 학습을 멈춥니다.")
      self.model.stop_training = True
```

```python
# 생성 (저장)
my_cb = myCallback()
```

```python
# callbacks= 파라미터 사용
model_7.fit( training_images, training_labels, epochs=40, callbacks=[ my_cb ] )
```

```python
model_7.evaluate(test_images, test_labels)
```
