## Stage 2: Import dependencies for the project

```python
import numpy as np
import datetime
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
```

```python
# 버전 확인
tf.__version__
```

## Stage 3: Dataset preprocessing



### Loading the dataset

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

### Image normalization



```python
# Feature Scaling
X_train = X_train / 255.0
```

```python
X_test = X_test / 255.0
```

### Reshaping of the dataset



```python
X_train.shape
```

```python
# 28 X 28 의 사진이 60000개 있음 (3차원 구성)
```

```python
type(X_train)
```

3차원을 2차원배열로 바꾸기  
( row , col ) 로 맞춰주고   
( -1, 2d * 3d) 로 하면 됨   
`(-1, 28 * 28)`

```python
X_train = X_train.reshape(-1, 28 * 28)
```

```python
X_train.shape
```

```python
X_test = X_test.reshape(-1, 28 * 28)
```

```python
X_test.shape
```

## Stage 4: Building an Artificial Neural network

### Defining the model



```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout
```

### Adding the first layer (Dense layer)

Layer hyper-parameters:
- number of units/neurons: 128
- activation function: ReLU
- input_shape: (784, )

```python
model = Sequential()
```

```python
model.add ( Dense( units=128, activation='relu', input_shape = (784, )) )
```

### Adding a Dropout layer 
오버 피팅을 막아주는 기능  
특정 w선을 잘라? drop out 해서 노드를 막아줌


```python
model.add( Dropout(rate=(0.2) ) )
```

### Adding the second layer (output layer)

- units == number of classes (10 in the case of Fashion MNIST)
- activation = 'softmax'

```python
model.add( Dense(units=10, activation='softmax' ) )
```

### Comiling the model

- Optimizer: Adam
- Loss: Sparse softmax (categorical) crossentropy 

```python
model.compile( optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

### Training the model

```python
model.fit(X_train, y_train, epochs=30)
```

```python
# 에포크는 30까지 셋팅하고, 정확도가 92% 이상이면 중지하도록 코딩
```

### Model evaluation and prediction

```python


class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if (logs.get('accuracy') > 0.92):
           print("\n정확도 0.92 이상이면 학습을 멈춥니다.")  
           self.model.stop_training = True
```

```python
# 객체 다시 생성 후 실행
```

```python
my_callback_fn = myCallback()
```

```python
model.fit(X_train, y_train, epochs=30, callbacks=[ my_callback_fn ])
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)
```
