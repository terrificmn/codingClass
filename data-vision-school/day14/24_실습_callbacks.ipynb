```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten
```

```python
(X1_train, y1_train), (X1_test, y1_test) = mnist.load_data()
X1_train = X1_train.reshape(-1, 28 * 28)
X1_test = X1_test.reshape(-1, 28 * 28)

# feature scaling
X1_train = X1_train / 255.0
X1_test = X1_test / 255.0

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

def build_classifier(optimizer) :
    classifier = Sequential()
    classifier.add( Flatten() )
    classifier.add ( Dense( units=256, activation='relu') )
    classifier.add ( Dense( units=512, activation='relu') )
    classifier.add ( Dense( units=700, activation='relu') )
    classifier.add ( Dense( units=10, activation='softmax') )
    classifier.compile(loss='sparse_categorical_crossentropy', optimizer = optimizer , metrics=['accuracy'] )
    
    return classifier

classifier = KerasClassifier(build_fn = build_classifier)

param = {
  'batch_size' : [500, 100],
  'epochs': [10],
  'optimizer': ['adam'] 
}

grid_search = GridSearchCV(estimator= classifier, param_grid=param, scoring='accuracy')
```

```python
result = grid_search.fit(X1_train, y1_train, verbose=1)
```

```python
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        i = 0.995
        if (logs.get('accuracy') > i ):
           print("\n정확도 {}가 되어서 학습을 종료합니다".format(i*100))  
           self.model.stop_training = True

myCallback = myCallback()
```

```python
# 함수로 작성.
def train_mnist():
  # MNist 데이터   
  (X_train, y_train), (X_test, y_test) = mnist.load_data()
  X_train = X_train.reshape(-1, 28 * 28)
  X_test = X_test.reshape(-1, 28 * 28)

  # feature scaling
  X_train = X_train / 255.0
  X_test = X_test / 255.0

  # modeling
  model = Sequential()
  model.add ( Flatten() )
  model.add ( Dense( units=256, activation='relu') )
  model.add ( Dense( units=512, activation='relu') )
  model.add ( Dense( units=700, activation='relu') )
  model.add ( Dense( units=10, activation='softmax') )

# add메소드를 사용 안하고 리스트로 넣어도 된다
'''
  model = tf.keras.models.Sequential ( [ 
    Flatten(),
    # units=를 생략하고 숫자만 써도 됨, 처음 파라미터로 써야함
    Dense(512 , activation= tf.nn.relu ),
    Dense(10 , activation= tf.nn.softmax ),
  ] )
'''

  model.compile( optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'] )

  model_history = model.fit(X_train, y_train, epochs=10, batch_size=500, callbacks=[myCallback ], verbose=1) 

  # model fitting
  return model_history.epoch, model_history.history['accuracy'][-1]
```

```python
result_epoch, result_accuracy = train_mnist()
```
