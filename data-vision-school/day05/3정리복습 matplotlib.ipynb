```python
# %로 차트 그려주는 것은 pie
# 다른 plot과는 다른게 data= 이 생략, 바로 DataFrame 이름을 넣어주면 됨

# labes = 
# autopct = '%f' ----> 퍼센트로 보여줌 f는 float 
# startangle= 90 -----> 시작할 앵글
# couterclock=False ---> 반시계방향으로 보여줌 , False면 시계방향
# wedgeprops= { 'width': 1.0 }  -----> 딕셔너리로 , 중간의 도넛구멍


# 먼저 한 컬럼을 시리즈로 만들기 (변수에 저장)
pokemon.head(3)
```

```python
GI_counts = pokemon.generation_id.value_counts()
```

pie (파라미터 많음)

```python
#pie는 옵션이 많음
# counterclock=False   ---> 반시계 방향 설정
# wedgeprops={ 'width': 0.7 }     --->  도넛구멍, 숫자가 1 이하로 작을 수록 커짐
# labels= DataFrame.index  ---->  라벨, 특정 컬럼의 인덱스를 사용하면 좋다
# autopct="%f"    ---> 숫자표기
# startangle=90   ---> 시작 앵글
# textprops={'color':"b"}  ---> 글자색 바꾸기 w: white, b: blue
plt.pie(GI_counts, labels= GI_counts.index, autopct="%f", startangle=90, counterclock=False, wedgeprops={ 'width': 0.7 }, textprops={'color':"w"} )
plt.show()
```

```python
# plt.hist()
# 히스토그램
# bins을 (한 구간)을 범위를 줘서 조절함 (빈의 갯수로 조절)
```

```python
pokemon.head(3)
```

```python
plt.hist(data=pokemon, x='speed', rwidth=0.8, bins = 20)
# rwidth=0.8 는 bin과 bin의 사이를 띄어준다
# bins= 20 빈의 갯수 설정
# 빈의 갯수로 보여지는게 왜곡 될 수 있음
plt.xlabel('speed')

plt.show()
```

```python
# pokemon DataFrame의 speed컬럼의 최소, 최대값으로 빈의 크기를 조절할 때 사용할 수 있음
# 먼저 통계치를 보고 
# 1차원 벡터를 만듬
pokemon.speed.describe()
```

```python
# 이런식으로 최소값, 최대값, 스텝값으로 np1차원배열을 만들어 주면
# step 파라미터가 bins의 크기가 된다
myBins = np.arange(5, 160, 20)
```

```python
plt.hist(data=pokemon, x='speed', rwidth=0.8, bins = myBins)
plt.show()
```

```python
# scatter 사용
# x, y축을 다 사용해서 만들 때 
# 두 컬럼의 상관관계를 알고 싶을 때 사용
# pd.corr() 같은 효과이면서 차트 그린 후 pd.corr()도 포함해 주면 좋다
# 행(row)의 데이터가 찍힘
```

```python
fuel = pd.read_csv('fuel_econ.csv')
```

```python
plt.scatter(data = fuel, x ='displ', y='comb')
plt.xlabel('DSIPLACEMENT L')
plt.ylabel('combined fuel eff')
plt.show()
```

```python
fuel[ ['displ', 'comb'] ].corr()
```

```python
# regplot 사용하기 
# scatter와 비슷하지만 중간에 선이 들어감
# regression (다른 변수간(컬럼)의 관계를 선으로 표현하기도 한다고 함)
```

```python
sb.regplot(data = fuel, x='displ', y='comb')
plt.show()
```

```python
#스캐터는 여러 데이터가 한 군데에 뭉쳐서 보이면 보기 힘듬
# 히트맵으로 해결
```

```python
# hist가 포함된 것이라 bins도 사용할 수 있음
plt.hist2d(data=fuel, x='displ', y='comb', cmin=0.5, bins=30)
plt.colorbar()
plt.show()
```
