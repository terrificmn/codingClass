Modeling

```python
여태껏 머신러닝은 알고리즘으로 (문제를 해결하기 위한 순서와 방법) 해결한 것이었음
딥러닝은 알고리즘으로 돌아가는 것이 아니고, 인공 신경망을 이용해서 수학식(방정식), 액티베이션 함수으로 
돌아가는 것
y = ax1 + bx2 + cx3 + e ( x는 컬럼의 값들이 들어가짐) Linear Regression에서 한 방정식
이 처럼 이 공식이 뉴런에 왼쪽의 공식에 들어가짐

여기에서는 11개 컬럼이 되어서 (One Hot Encoding으로, Label Encoding)
input layer의 뉴런이 11개가 되는셈 (컬럼 수가 결정)

ax1 + bx2 + cx3 + dx4 .... kx11 + q
컴퓨터는 여기에서는 a, b, c... x 값을 찾는 것 
인공신경망 그림에서는 선 들을 의미한다 == w 라고 하고 == weight == 가중치 라고 함
```

텐서플로우 불러오기

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

```python
Gradient Descent
기울기는 미분
기울기를 구했으면 마이너스를 해준다. 그러면 내리가는 것 +면 올라가는 것
한발 내려가고 오차계산하고  (여기서 한발 내려간다는 의미는 )
입력에서 (output)까지 계산한 것을 다시 w를 다 바꿔주면서 (update)를 해준 다음에
입력까지 다시 와서 다시 오차 계산을 해주는 것을 (한발 내려간것)

Stochastic Gradient Descent (sgd) 라고 함
```

Sequential()생성

```python
classifier = Sequential()
```

```python
X_scaled.shape
# Input layer의 입력 뉴런이 11개 된다 (컬럼 수와 동일)
```

```python
# layer는 Dense라고 하며 Dense()를 사용
```

.add (Dense ()) 레이어 추가

```python
# units = layer 들을 의미 
# input layer 옆에 있는 layer
classifier.add( Dense( input_dim = 11, units = 6, activation='relu') )
```

```python
# Vanishing Gradient 현상
# Activation Functions를 Sigmoid, Tanh를 hidden layer에서 설정해놓으면
# Sigmoid, Tanh가 0이 되버리면 오차계산이 다 안 끝났는데 계산이 끝나버리면서 
# 오차가 생겨버리는 현상 
# 그래서 Sigmoid, Tanh 함수는 히든레이어에서는 사용하지 않는다

# activation=tf.nn.relu 는 액티베이션 함수 지정인데 위의 
# activation='relu' 와 같은 의미
classifier.add( Dense( units= 8, activation=tf.nn.relu) )
```

```python
# 마지막 outlayer 설정 // 마지막에는 sigmoid 함수를 사용할 수 있다
classifier.add( Dense( units=1, activation=tf.nn.sigmoid) ) #같은 표현 == activation='sigmoid'
```

```python
# complie => loss 함수 (오차 함수)를 무엇으로 할지, 옵티마이저는 무엇으로 할지 결정
```

complie 설정

```python
# loss= 오차 계산하는 방식 'binary_crossentropy' 
# optimizer='adam' 잘 모르겠으면 adam 함수, 
# 참고로: sgd 도 있었음 Stochastic Gradient Descent
```

```python
classifier.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'] )
```

학습

```python
# epochs --> 한정된 데이터를 가지고 다시 입력을 다시 해서 w가중치를 계속바꿀 수 있으므로
# 다시 반복해서 학습하게 하는 것 (오차가 줄어듬)
# batch_size=10 ---> 행렬로 한꺼번에 대입해서 계산하게 함

# ** 중요 두 번 실행하면 학습을 했던 데이터를 가지고 다시 학습 하기 때문에 
# 파라미터를 바꿔서 할 때는 처음 객체 생성부터 다시 실행 시켜줘야한다
classifier.fit(X_train, y_train, epochs=20, batch_size=10)
```

### 용어 정리

epoch

- 한 번의 epoch는 신경망에서 전체 데이터 셋에 대해 forward pass/backward pass 과정을 거친 것을 말함. 즉, 전체 데이터 셋에 대해 한 번 학습을 완료한 상태


batch_size

메모리의 한계와 속도 저하 때문에 대부분의 경우에는 한 번의 epoch에서 모든 데이터를 한꺼번에 집어넣을 수는 없습니다. 그래서 데이터를 나누어서 주게 되는데 이때 몇 번 나누어서 주는가를 iteration, 각 iteration마다 주는 데이터 사이즈를 batch size라고 합니다.

출처: https://www.slideshare.net/w0ong/ss-82372826

```python
# Tuning the ANN
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense
```

```python
#모델링이면 
#1. Sequential
#2. Dense 
#3. complie 
# 순서로 만듬
```

```python
# 위의 순서를 사용자 함수로 만들기
def build_classifier(optimizer) :
  classifier = Sequential()
  classifier.add( Dense( input_dim = 11, units = 6, activation='relu') )
  classifier.add( Dense( units= 8, activation=tf.nn.relu) )
  classifier.add( Dense( units=1, activation=tf.nn.sigmoid) )
  classifier.compile(loss='binary_crossentropy', optimizer = optimizer , metrics=['accuracy'] )
  return classifier
```

KerasClassifier 생성

```python
# build_fn= 파라미터
# 사용자 함수를 지정할 수 있음
classifier = KerasClassifier(build_fn = build_classifier)
```

```python
param = {
    'batch_size' : [10, 20, 25],
    'epochs': [200, 500, 700],
    'optimizer': ['adam', 'rmsprop'] 
}
```

```python
# scoring='accuray' 대개 정확도를 선택
# cv= 원래 있던 트렌인셋을 10개로 쪼개서 더 정교하게 학습을 시키는 것
# 부분 돌아가면서 학습을 함,
# cross validation
grid_search = GridSearchCV(estimator= classifier, param_grid=param, scoring='accuracy', cv=10)
```

```python
grid_search.fit(X_train, y_train)
```
