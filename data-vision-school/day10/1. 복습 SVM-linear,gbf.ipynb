```python
cancer = load_breast_cancer()
```

```python
# np.c_ 이용해서 데이터 합치기
# 이렇게 데이터를 합치는 것은 DataFrame을 만들 때 data= 영역에 들어가지게 됨
np.c_[ cancer['data'], cancer['target'] ]
```

```python
# np.append()로 feature_names 리스트에 , 'target' 이름 추가
# 이제 이 놈들은 columns = 에 들어가면 됨
np.append( cancer['feature_names'], ['target'])
```

```python
# 이제 pd.DataFrame으로 만들기
# np.c 로 데이터 합치기
# np.append()
df = pd.DataFrame(data = np.c_[ cancer['data'], cancer['target'] ], columns= np.append( cancer['feature_names'], ['target']) )
```

pairplot 으로 모든 컬럼 해보기

```python
# sb.pairplot()을 할 때 
# 컬럼명으로 vars= 파라미터 넘겨줄 수 있음
df.columns
```

```python
cols = ['mean radius', 'mean texture', 'mean perimeter', 'mean area']
```

```python
#여러개의 컬럼간의 관계 비교하기
# hue='컬럼명' ----> 색깔을 구분지어 줄 수 있음
# hue='target'은 0과 1 인데 악성, 양성을 의미함 , 색으로도 구분지어 볼 수 있음
sb.pairplot( data= df, vars= cols, hue='target')
plt.show()
```

```python
# sb.countplot() 특정컬럼의 수를 카운트해서 차트 그림
sb.countplot(data=df, x='target')
plt.show()
```

```python
# 히트맵으로 상관관계 숫자 차트그리기
# sb.heatmap(data=) ---> 여기서 data= df.corr() 로 파라미터를 넘겨주면 상관관계 분석
plt.figure(figsize=(15,15))
sb.heatmap(data=df.corr(), annot=True)
plt.show()
```

```python
# X,  y 로 데이터 나누기
```

```python
df.head()
```

```python
X = df.iloc[:, 0: -2 +1]
```

```python
y = df.iloc[:, -1:]
```

```python
#evaluating the model
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
X = mm.fit_transform(X)
```

```python
# train_test_split 하기
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=777)
```

```python
# SVC linear 와 SVC rbf 로 학습
```

```python
from sklearn.svm import SVC
```

```python
# SVC( kernel='linear') 방식으로 생성
classifier_l = SVC(kernel='linear')
```

```python
classifier_l.fit(X_train, y_train)
```

```python
y_pred_l = classifier_l.predict(X_test)
```

```python
classifier_r = SVC(kernel='rbf')
```

```python
classifier_r.fit(X_train, y_train)
```

```python
y_pred_r = classifier_r.predict(X_test)
```

```python
# 2개의 학습 방식 비교, SVC linear / rbf
```

```python
y_pred_l
```

```python
y_pred_r
```

```python
# confusion_matrix 로 정확도 확인
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm_l = confusion_matrix(y_test, y_pred_l)
```

```python
cm_l
```

```python
(35+75) / cm_l.sum()
```

```python
cm_r = confusion_matrix(y_test, y_pred_r)
```

```python
cm_r
```

```python
(35+74) / cm_r.sum()
```

Grid Search

```python
#SVC 의 rbf 의 하이퍼 파라미터는 (사람이 정할 수 있는 파라미터)
# 2개가 있음
# gamma 와 C
# 여러번 반복해서 가장 좋은 gamma, c의 파라미터의 가장 좋은 값을 돌려준다
```

from sklearn.model_selection import GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
# 먼저 딕셔너리 형태로 만들어 줘야함
param = {
    'kernel': ['rbf'],
    'gamma': [0.001, 0.01, 0.1, 1, 10],
    'C': [0.1, 1, 10, 50, 100]
 
}
```

```python
# 인스턴스 생성
```

GridSerachCt (SVC(), refit=True 로 함
위의 param 딕셔너리에 저장한 kernel, gamma, C 등의 파라미터도 넘김

```python
grid= GridSearchCV( SVC(), param, refit=True, verbose =4 )
# verbose= 파라미터는 화면에 보여주는 것
```

```python
#학습
# 위에서 지정한 파라미터 값대로 반복해서 수행한다
grid.fit(X_train, y_train)
```

가장 좋은 조합 보여주기 
grid.best_params_

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.best_estimator_
```

```python
y_pred_grid  = grid.predict(X_test)
```

```python
cm_grid = confusion_matrix(y_test, y_pred_grid)
```

```python
cm_grid
```

```python
(35+74) / cm_grid.sum()
```

```python
sb.heatmap(data=cm_grid, annot=True)
plt.axis('off')
plt.show()
```
