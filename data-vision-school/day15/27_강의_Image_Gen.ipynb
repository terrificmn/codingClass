## Building a Small Model from Scratch

간단한 모델링 하기.
사진의 결과는 2개중의 하나이미로, 맨 마지막 액티베이션 함수는 시그모이드 사용

```
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Flatten
```

```
from keras.layers import Conv2D, MaxPooling2D
```

```
edu = Sequential()
edu.add (Conv2D (16, (3, 3), activation='relu', input_shape= (300, 300, 3) ))
edu.add (MaxPooling2D(2, 2) )
edu.add (Conv2D (32, (3, 3), activation='relu')) 
edu.add (MaxPooling2D(2, 2) )
edu.add (Conv2D (64, (3, 3), activation='relu')) 
edu.add (MaxPooling2D(2, 2) )
edu.add (Conv2D (64, (3, 3), activation='relu')) 
edu.add (MaxPooling2D(2, 2) )
edu.add (Conv2D (64, (3, 3), activation='relu')) 
edu.add (MaxPooling2D(2, 2) )
edu.add (Flatten() )
edu.add (Dense(512, activation='relu'))
edu.add(Dense(1, activation='sigmoid'))
```

```
edu.summary()
```

```
# 모델링
```

```
## 내가 짠 모델은 잘 맞추지 못함

model = Sequential( [
  Conv2D( filters = 128, kernel_size=(6, 6), activation='relu', input_shape= (300, 300, 3) ) ,
  MaxPooling2D( 2, 2 ),
  Conv2D( filters = 64, kernel_size=(4, 4), activation='relu' ),
  MaxPooling2D( 2, 2),
  Conv2D( filters = 32, kernel_size=(3, 3), activation='relu' ),
  MaxPooling2D( 2, 2),
  Flatten(),
  Dense(521, activation='relu'),
  Dense(128, activation='relu'),
  Dense(1, activation='sigmoid')
]) 
```

```
model = Sequential( [
  Conv2D( filters = 64, kernel_size=(3, 3), activation='relu', input_shape= (300, 300, 3) ) ,
  MaxPooling2D( 2, 2 ),
  Conv2D( filters = 64, kernel_size=(3, 3), activation='relu' ) ,
  MaxPooling2D( 2, 2 ),
  Conv2D( filters = 32, kernel_size=(3, 3), activation='relu' ),
  MaxPooling2D( 2, 2),
  
  Conv2D( filters = 32, kernel_size=(3, 3), activation='relu' ),
  MaxPooling2D( 2, 2),
  
  Conv2D( filters = 32, kernel_size=(3, 3), activation='relu' ),
  MaxPooling2D( 2, 2),
  Conv2D( filters = 32, kernel_size=(3, 3), activation='relu' ),
  MaxPooling2D( 2, 2),
  Flatten(),
  Dense(128, activation='relu'),
  Dense(64, activation='relu'),
  Dense(1, activation='sigmoid')
  ])  
```

```
model.summary()
```

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```
# 이미지를 피쳐 스케일링 한다. 255.0으로 나누기
```

```
# 255로 나누것과 같은 것 / 소수점으로 나눠준다
train_datagen = ImageDataGenerator(rescale= 1/255.0)
```

```
# 이미지가 들어있는 디렉토리 정보를 주고, 사이즈 정보도 준고, 분류의 정보도 준다
train_generator = train_datagen.flow_from_directory( 
    # 파일이 위치한 상위디렉토리를 적어준다
    '/tmp/horse-or-human', 
    # 타켓사이즈는 원본대로 할 수도, 마음대로 바꿔도 됨
    target_size=( 300, 300 ), #현재 원본 사이즈
    class_mode='binary' #2개로 분류 (말/사람)
)

# 위의 train_datagen.flow_from_directory() 는 파일을 찾아온다
```

```
# 트레이닝 셋
validation_train_datagen = ImageDataGenerator(rescale= 1/255.0)
```

```
validation_train_generator = validation_train_datagen.flow_from_directory( 
    # 파일이 위치한 상위디렉토리를 적어준다
    '/tmp/validation-horse-or-human', 
    # 타켓사이즈는 원본대로 할 수도, 마음대로 바꿔도 됨
    target_size=( 300, 300 ), #현재 원본 사이즈
    class_mode='binary' #2개로 분류 (말/사람)
)
```

prints a summary of the NN 

**NOTE**: 각 알고리즘 레퍼런스 페이지 

 [RMSprop optimization algorithm](https://wikipedia.org/wiki/Stochastic_gradient_descent#RMSProp) RMSprop automates learning-rate tuning for us

 [stochastic gradient descent](https://developers.google.com/machine-learning/glossary/#SGD) (SGD)
 
 
  [Adam](https://wikipedia.org/wiki/Stochastic_gradient_descent#Adam) and [Adagrad](https://developers.google.com/machine-learning/glossary/#AdaGrad), also automatically adapt the learning rate during training, and would work equally well here.

 RMSprop optimization algorithm 사용하여 컴파일 해보자

```
from tensorflow.keras.optimizers import RMSprop
# optimizer 함수 오차를 줄이는 방법 : adam, sgd, etc...
model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(lr=0.001),
              metrics=['accuracy'])
```

### Data Preprocessing



```
edu.compile(loss='binary_crossentropy',
              optimizer=RMSprop(lr=0.001),
              metrics=['accuracy'])
```

### Training
Let's train for 15 epochs

```
epoch_histroy_edu = edu.fit(train_generator, steps_per_epoch = 8, epochs =15, verbose =1 )
```

###Running the Model




```
# 트레이닝 셋
```

```
# validation_split = 0.2 은 테스트 (validate)할 데이터가 없을 경우에 하는 것
# validataion_data는 위에서 이미지 파일 노말라이즈 해준 파일로 해줌
validation_epoch_histroy = model.fit(validation_train_generator, steps_per_epoch = 8, epochs =15, verbose =1 , validation_data = validation_train_generator)
```

```
# validation 으로 학습한 것 정확도 val_accuracy 보기
```

```
plt.plot(validation_epoch_histroy.history['accuracy'], color = 'b')
plt.plot(validation_epoch_histroy.history['val_accuracy'], color ='r')
plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['Train', 'Val']) #Val은 training셋에서 자체 25%테스트 한 것
plt.show()
```

```
# validation 으로 학습한 것 오차와  val_accuracy 보기
```

```
plt.plot(validation_epoch_histroy.history['loss'], color = 'b')
plt.plot(validation_epoch_histroy.history['val_loss'], color ='r')
plt.title('model loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['Train', 'Val']) #Val은 training셋에서 자체 25%테스트 한 와
plt.show()
```
