## Training the Apriori model on the dataset

```python
from apyori import apriori
# apyori는 파이썬 패키지
# apriori는 알고리즘 모듈
```

```python
# data프레임의 numpy array를 넣어줘야함 --> df.values
# 어떤 물품을 사면 같이 들어있는 경우
# min_support는 예를 들어 햄버거와 프렌치 프라이가 같이 있을 경우
# min_lift = 3 3이상인 것을 찾음 (1이면 관련이 없는 것)

apriori( transactions=df.values , min_support = 0.002, min_confidence = 0.3, min_lift = 3, min_length= 2, max_length=2)
# 1000개중에 3개가 있으면 데이터를 뽑아라
```

```python
rules = apriori( transactions=df.values , min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length= 2, max_length=2)
```

## Visualising the results

```python
# 리스트화 해줌
results = list(rules)
```

```python
results
```

### Putting the results well organised into a Pandas DataFrame

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))

```

```python
df_result = pd.DataFrame( inspect(results), columns=[ 'Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift' ])
```

### Displaying the results non sorted

```python
df_result
# Left Hand Side를 사면 Right Hand Side를 사게 되는 것
# lift가 높을수록 높은 관계
```

### Displaying the results sorted by descending lifts

```python
df_result.sort_values(by='Lift', ascending=False)
```
